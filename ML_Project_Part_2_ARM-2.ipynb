```r
df_usa_transactions <- read.transactions("/content/merged_usa_dataset.csv",
                             rm.duplicates = FALSE,
                             format = "basket",
                             sep=",",
                             cols=NULL)
inspect(df_usa_transactions)
```

```r
# Specify multiple LHS values
lhs_values = c("Electricity from wind (TWh)", "Geo Biomass Other - TWh", "Geothermal Capacity", "Hydro Generation - TWh", "Other renewables including bioenergy (TWh)", "Solar Capacity","Solar Generation - TWh","Wind Capacity", "Wind Generation - TWh")

# Use the lhs parameter with the specified LHS values
Rules <- apriori(data = df_usa_transactions,
                     parameter = list(supp = 0.01, conf = 0.05, minlen = 2),
                     appearance = list(default = "rhs",lhs = lhs_values),
                     control = list(verbose = FALSE))

inspect(Rules[1:5])
```

```r
# Top 15 rules by support
Rules <- sort(Rules, decreasing = TRUE, by = "support")
inspect(Rules[1:15])
```

```r
# Top 15 rules by confidence
Rules <- sort(Rules, decreasing = TRUE, by = "confidence")
inspect(Rules[1:15])
```

```r
# Top 15 rules by lift
Rules <- sort(Rules, decreasing = TRUE, by = "lift")
top_15_rules = Rules[1:15]
inspect(Rules[1:15])
```
