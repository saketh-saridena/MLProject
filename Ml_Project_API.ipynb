```python
import requests
import pandas as pd

# Define the API URL
url = "https://api.eia.gov/v2/total-energy/data/?api_key=3OHnrJ1x9pyZxd0DdAkVXSgE4iCdkOW1mySHXlwf&frequency=annual&data[0]=value&start=2020&end=2023&sort[0][column]=period&sort[0][direction]=desc&offset=0&length=5000"

# Define the headers
headers = {'X-Params': '{"frequency": "annual", "data": ["value"], "facets": {}, "start": "2020", "end": "2023", "sort": [{"column": "period","direction": "desc"}],"offset": 0,"length": 5000}'}

# Send a GET request to the EIA API
response = requests.get(url, headers=headers)

# Check the status of the request
if response.status_code == 200:
    # Parse the response JSON
    data = response.json()

    # Extract the series data
    series_data = data['response']['data']

    # Convert the data to a pandas DataFrame
    df = pd.DataFrame(series_data)

    # Convert the 'period' column to datetime format
    df['period'] = pd.to_datetime(df['period'])

    # Set the 'period' column as the index
    df.set_index('period', inplace=True)

    # Sort the DataFrame by date
    df.sort_index(inplace=True)

    print(df)
else:
    print(f'Request failed with status code {response.status_code}')
```
